#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python

precio_max = df['price'].idxmax() #Se busca dentro de la columna price el precio mayor 
datos_completos = df.loc[precio_max] #Se extraen los datos completos de la fila de la casa con el precio mayor
print(f"La casa en {datos_completos['address']}, es la mas cara y su precio es de {datos_completos['price']} USD") #Se muestra la direccion y el precio de la casa que se ha buscado
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python

datos_sin_cero = df[df['price']>0] #Se eliminan todas las propiedad cuyo valor es > 0
precio_min = datos_sin_cero['price'].idxmin() #Se busca dentro de la columna price el precio menor
datos_completos_min = df.loc[precio_min] #Se extraen los datos completos de la fila de la casa con el precio menor
print(f"La casa en {datos_completos_min['address']}, es la mas barata y su precio es de {datos_completos_min['price']} USD") #Se muestra la direccion y el precio de la casa que se ha buscado
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python

mas_grande = df['surface'].idxmax() #Se busca en la columna surface la casa mas grande
completos_mas_grande = df.loc[mas_grande] #Se extraen los datos de la fila completa de la casa mas grande
print(f"La casa mas grande esta ubicada en {completos_mas_grande['address']} y su superficie es de {completos_mas_grande['surface']} metros") #Se muestra por pantalla la ubicacion de la casa mas grande y su superficie

mas_peque = df['surface'].idxmin() #Se busca en la columna surface la casa mas pequeña
completos_mas_peque = df.loc[mas_peque] #Se extraen los datos de la fila completa de la casa mas pequeña
print(f"La casa mas pequeña esta ubicada en {completos_mas_peque['address']} y su superficie es de {completos_mas_peque['surface']} metros") #Se muestra por pantalla la ubicacion de la casa mas pequeña y su superficie
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

Cuenta el número de poblaciones únicas en la columna 'level5' e imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3, ...`

```python

columna = df['level5'] #Se extraen todos los datos de la columna level 5
extraccion = columna.unique() #Se transforman a datos unicos para no mostrar nombres repetidos
extraccion_sin_na = extraccion.dropna() #Se eliminan en caso de que existan los NA para evitar error
print(','.join(extraccion_sin_na)) #Se muestran por pantalla separados por coma los datos extraidos 
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.


```python

print(f'El tamaño del dataframe con valores nulos es: {len(df)}') #Mostramos el tamaño del dataframe sin modificar los valores nulos

sin_filas_nulas = df.dropna(how='all') #Eliminamos la fila solo si todos los valores de la fila son nulos

print(f'El tamaño del dataframe con algunos valores nulos, pero no filas completas, es: {len(sin_filas_nulas)}') #Se muestra el dataframe ya modificado pero como no hay filas enteras de nulos queda igual

filas_null = df.dropna() #Se eliminan las filas en las que hay al menos un valor nulo

print(f'El tamaño del dataframe sin valores nulos es: {len(filas_null)}') #Se muestra el dataframe modificado 


```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
localidad = df[df['level5']=='Arroyomolinos (Madrid)'] #Extraer por la columna level 5 solo los que son de Arroyomolinos 

precios = int(localidad['price'].mean())  #Se saca la media aritmetica de la columna price
print(precios) #Se muestra por pantalla el resultado
```
